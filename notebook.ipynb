```python
# prompt: I want a code that can generate text in the format, 4 letters then 4 numbers like this "AAAA0008"
import numpy as np
import random
import string

np.random.seed(2025)

def generate_code():
  letters = ''.join(random.choices(string.ascii_uppercase, k=4))
  numbers = ''.join(random.choices(string.digits, k=4))
  return letters + numbers

# Example usage:
code = generate_code()
code
```

```python
amala_id_list = []
for i in range(50):
  amala_id_list.append(generate_code())
```
